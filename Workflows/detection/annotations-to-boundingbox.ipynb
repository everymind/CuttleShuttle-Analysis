```python
from __future__ import print_function, division, absolute_import, unicode_literals
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os
import cv2
%matplotlib inline
home = os.path.expanduser('~/')
```

```python
csv_path = os.path.join(home, 
                        'CuttleShuttle-Analysis/Workflows/ShaderNavigator/annotations/L1-H2013-03',
                        'session_video2014-09-16T14_45_26.csv')
```

```python
df=pd.read_csv(csv_path, 
               delim_whitespace=True,
               usecols=range(4),
               header=None)
data=df.values
print(data[0,1])
```

```python
cv_img = cv2.imread(os.path.join(home, 
                        'cuttle-selfies/L1-H2013-03', 
                        'session_video2014-09-16T14_45_26_f0004784.png'))
```

```python
cv_img_for_draw = np.copy(cv_img)
```

```python
transform_data_cordinate = np.empty((data.shape[0],2),dtype=np.int16) #placeholder for coordinates
transform_data_cordinate[:,0] = (data[:,-2]+1)*cv_img.shape[1]//2 
transform_data_cordinate[:,1] = (data[:,-1]+1)*cv_img.shape[0]//2
```

```python
transform_data_cordinate[:,1] = cv_img.shape[0]-transform_data_cordinate[:,1]
```

```python
cv2.circle(cv_img_for_draw, tuple(transform_data_cordinate[0,:]), radius=10, color=(255,0,0), thickness=4)
cv2.circle(cv_img_for_draw, tuple(transform_data_cordinate[1,:]), radius=10, color=(255,0,0), thickness=4)
plt.figure(figsize=(18,10))
plt.imshow(cv_img_for_draw)
```
